## Задание 1

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rc('legend', fontsize=15)
```

```python
def f(x):
    return np.sin(x / 5) * np.e ** (x / 10) + 5 * np.e ** (- 1 / 2 * x)
```

```python
def make_poly_approx(n):
    size = n
    x = np.linspace(1, 15, size)
    power_array = np.linspace(0, size - 1, size)
    A = np.power((np.ones((size, size)) * x).T, power_array)
    B = f(x)
    W = np.linalg.solve(A, B)
    return np.sum((A * W), axis=1)
```

```python
fig, axes = plt.subplots(2, figsize=(15,11))
n = 100
x = np.linspace(1, 15, n)
axes[0].plot(x, f(x), label="exact solution")
axes[1].plot(x, f(x), label="exact solution")
n = 5
x = np.linspace(1, 15, n)
y = make_poly_approx(n)
axes[0].plot(x, y, label=f"polynomial solution, n = {n}")
axes[0].plot(x, y, "ro")
axes[0].legend()
n = 15
x = np.linspace(1, 15, n)
y = make_poly_approx(n)
axes[1].plot(x, y, label=f"polynomial solution, n = {n}")
axes[1].plot(x, y, "ro")
axes[1].legend()
```

### Аппроксимация плохо приближает решение при n=5, лучше приближает при n=15. Чем больше количество точек, тем лучше работает приближение

## Задание 2

```python
def make_poly_approx_with_error(n):
    size = n
    error = np.random.normal(0, 0.01, size)
    x = np.linspace(1, 15, size)
    power_array = np.linspace(0, size - 1, size)
    A = np.power((np.ones((size, size)) * x).T, power_array)
    B = f(x) + error
    W = np.linalg.solve(A, B)
    return np.sum((A * W), axis=1)
```

```python
fig, axes = plt.subplots(3, figsize=(15,15))
n = 100
x = np.linspace(1, 15, n)
axes[0].plot(x, f(x), label="exact solution")
axes[1].plot(x, f(x), label="exact solution")
axes[2].plot(x, f(x), label="exact solution")
n = 5
x = np.linspace(1, 15, n)
y = make_poly_approx_with_error(n)
axes[0].plot(x, y, label=f"polynomial solution with error, n = {n}")
axes[0].plot(x, y, "ro")
axes[0].legend()
n = 15
x = np.linspace(1, 15, n)
y = make_poly_approx_with_error(n)
axes[1].plot(x, y, label=f"polynomial solution with error, n = {n}")
axes[1].plot(x, y, "ro")
axes[1].legend()
n = 25
x = np.linspace(1, 15, n)
y = make_poly_approx_with_error(n)
axes[2].plot(x, y, label=f"polynomial solution with error, n = {n}")
axes[2].plot(x, y, "ro")
axes[2].legend()
```

### При небольшом количестве точек решение остаётся точным, даже при том, что присутствует шум. Но при увеличении количества точек, аппроксимация отклоняется от точного решения сильнее. Подозреваю, что это происходит из-за того, что количество неизестных W увеличивается и уже сумма частей полинома накапливает большую ошибку, так как каждый элемент W найден с небольшой ошибкой

## Задание 3

```python
def count_min_square_method(N):
    n = 3
    error = np.random.normal(0, 0.1, N)
    x = np.linspace(1, 15, N)
    power_array = np.linspace(0, n, n + 1)
    A = np.power((np.ones((n + 1, N)) * x).T, power_array)
    B = f(x) + error
    W = np.linalg.lstsq(A, B)
    return (np.sum((A * W[0]), axis=1)), B
```

```python
fig, axes = plt.subplots(2, figsize=(15,11))
n = 100
x = np.linspace(1, 15, n)
axes[0].plot(x, f(x), label="exact solution")
axes[1].plot(x, f(x), label="exact solution")
x = np.linspace(1, 15, n)
y = count_min_square_method(n)
axes[0].plot(x, y[0], label=f"min square solution, n = {n}")
axes[1].plot(x, y[1], "ro", label="polynomial points")
axes[0].legend()
axes[1].legend()
```

### Метод наименьших квадратов хорошо приближает решение, даже при том, что полином 3 степени. Методы из пунктов 1 и 2 такой точности при такой степени полинома предоставить не могут
