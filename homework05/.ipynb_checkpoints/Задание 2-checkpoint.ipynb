1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!
2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)
3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>!
```python
x = np.linspace(-1, 1, 250)
y = x ** 2 * np.sin(1 / (x ** 2)) + x
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
```

```python
x = np.linspace(5, 25, 100)
y = abs(1 - (1 / (1 + x ** 2)) / (1 / x ** 2))
plt.semilogy(x, y)
```

```python
x = np.linspace(1, 10, 150)
plt.figure(figsize=(18, 9))
plt.plot(x, scipy.special.j0(x), label='Функция Бесселя первого рода порядка 0')
plt.plot(x, scipy.special.y1(x), label='Функция Бесселя второго рода порядка 1')
plt.plot(x, scipy.special.itstruve0(x), label='Интеграл от функции Струве порядка 0')
plt.legend(fontsize = 20)
```

```python
from fractions import Fraction
plt.figure(figsize=(18, 9))
x = np.arange(-6 * np.pi, 6 * np.pi, np.pi / 1000)
y = np.sin(x) / x
plt.plot(x, y)
arr = np.arange(-6 * np.pi, 6.01 * np.pi, np.pi / 2)
plt.xticks(arr, list(map(lambda x: x + "π", map(str, np.arange(-6, 6.01, 0.5)))), fontsize=13)
plt.grid()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.sin(Y)

fig, ax = plt.subplots(figsize=(8, 7))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap=mpl.cm.Paired)
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X,Y = np.meshgrid(x,y)

Z = np.e ** (-X ** 2 - Y ** 2)

fig, ax = plt.subplots(figsize=(8, 7))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap=mpl.cm.Paired)
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```
